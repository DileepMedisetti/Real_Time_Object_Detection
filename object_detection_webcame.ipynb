```python
# Notebook-friendly webcam display (works without cv2 GUI)
# It captures frames, runs YOLO12 inference, and displays in the notebook using IPython.display.
# Stop the cell manually (stop button) or set a time limit below.
from ultralytics import YOLO
import cv2, time
from IPython.display import display, clear_output
from PIL import Image
import numpy as np

model = YOLO("yolo12n.pt")  # auto-download if needed
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError("Could not open webcam. Try different camera index (0,1,2).")

max_seconds = 30  # auto-stop after 30 seconds; set to None to run until manual interrupt
start = time.time()
frame_count = 0

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break
        frame_count += 1

        # Run inference
        results = model(frame, verbose=False)
        annotated_rgb = results[0].plot()  # RGB uint8
        # Convert to BGR for consistency, then to RGB PIL
        annotated_bgr = cv2.cvtColor(annotated_rgb, cv2.COLOR_RGB2BGR)
        img = Image.fromarray(cv2.cvtColor(annotated_bgr, cv2.COLOR_BGR2RGB))

        clear_output(wait=True)
        display(img)

        # Optional: break after max_seconds
        if max_seconds and (time.time() - start) > max_seconds:
            break

except KeyboardInterrupt:
    print("Stopped by user")
finally:
    cap.release()
    clear_output(wait=True)
    print("Webcam released.")
```

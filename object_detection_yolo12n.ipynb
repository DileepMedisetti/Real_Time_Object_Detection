```python
from ultralytics import YOLO
import cv2
from pathlib import Path

model = YOLO('yolo12n.pt')  # change to yolo26s.pt or custom checkpoint for better accuracy
print(model)
```

```python
# Run inference on an example image
res = model('https://ultralytics.com/images/zidane.jpg')
display_image = res[0].plot()  # RGB image
# Save annotated image
import matplotlib.pyplot as plt
plt.imshow(display_image)
plt.axis('off')
```

```python
# Example: Access raw boxes
boxes = res[0].boxes
if boxes is not None:
    print('xyxy:', boxes.xyxy)
    print('conf:', boxes.conf)
    print('cls:', boxes.cls)
```
